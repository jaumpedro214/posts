```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
```

```python
def normalize_names(name):
    return name.lower().strip().replace(' ', '-')
```

```python
df_results = pd.read_json('results.json')
df_results['dummy'] = df_results['predictions'].apply(lambda x: x['dummy']).apply(normalize_names)
df_results['gpt'] = df_results['predictions'].apply(lambda x: x['gpt'] ).apply(normalize_names)
df_results['gpt-structured'] = df_results['predictions'].apply(lambda x: x['gpt-structured']).apply(normalize_names)
df_results['rag'] = df_results['predictions'].apply(lambda x: x['rag']).apply(normalize_names)
```

```python
dummy_acc_score = 100*accuracy_score(df_results['true_artist'], df_results['dummy'])
gpt_acc_score   = 100*accuracy_score(df_results['true_artist'], df_results['gpt'])
gpt_structured_acc_score = 100*accuracy_score(df_results['true_artist'], df_results['gpt-structured'])
rag_acc_score = 100*accuracy_score(df_results['true_artist'], df_results['rag'])
```

```python
print(f"Dummy Model Accuracy: {dummy_acc_score:.2f}%")
print(f"GPT Model Accuracy: {gpt_acc_score:.2f}%")
print(f"GPT Structured Model Accuracy: {gpt_structured_acc_score:.2f}%")
print(f"RAG Model Accuracy: {rag_acc_score:.2f}%")
```
